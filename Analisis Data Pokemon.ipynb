```python
# Cleaning Data

# menghapus kolom indeks ke 0 dengan nama kolom '#'
# (Karena terjadi double id pada auto increment dan kolom '#' dari penampilan data)
data = data.drop(data.columns[0], axis = 1)
data
```

```python
# menambahkan kolom untuk menghitung jumlah tipe 
data['Type Count'] = data.apply(lambda row: 1 if pd.isna(row['Type 2']) else 2, axis=1)

# menghitung jumlah Pokemon berdasarkan jumlah tipe 
type_counts = data['Type Count'].value_counts()

# menampilkan hasil
print("Jumlah Pokemon dengan 1 tipe:", type_counts.get(1, 0))
print("Jumlah Pokemon dengan 2 tipe:", type_counts.get(2, 0))

# membuat pie chart
plt.figure(figsize=(8, 6))
plt.pie(type_counts, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])

# Menambahkan keterangan
plt.legend(labels=['2 Tipe', '1 Tipe'], title='Jumlah Tipe', loc='upper left', bbox_to_anchor=(1, 1))

plt.title('Persentase Pokemon Berdasarkan Jumlah Tipe')
plt.axis('equal')
plt.show()
```

```python
# menghitung jumlah Pokemon untuk setiap tipe elemen di kolom 'Type 1'
type1_counts = data['Type 1'].value_counts()

# mengambil 5 tipe elemen teratas
top_5_types = type1_counts.head(5)

# menampilkan hasil
print("Top 5 Tipe Elemen di 'Type 1':")
print(top_5_types)

# membuat bar chart untuk top 5 tipe elemen
plt.figure(figsize=(10, 6))
top_5_types.plot(kind='bar', color='skyblue')
plt.title('Top 5 Tipe Elemen Pokemon Pada Tipe 1')
plt.xlabel('Tipe Elemen')
plt.ylabel('Jumlah Pokemon')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# menghitung jumlah Pokemon untuk setiap tipe elemen di kolom 'Type 1'
type1_counts = data['Type 2'].value_counts()

# mengambil 5 tipe elemen teratas
top_5_types = type1_counts.head(5)

# menampilkan hasil
print("Top 5 Tipe Elemen di 'Type 2':")
print(top_5_types)

# membuat bar chart untuk top 5 tipe elemen
plt.figure(figsize=(10, 6))
top_5_types.plot(kind='bar', color='skyblue')
plt.title('Top 5 Tipe Elemen Pokemon Pada Tipe 2')
plt.xlabel('Tipe Elemen')
plt.ylabel('Jumlah Pokemon')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# menghitung jumlah PokÃ©mon untuk setiap generasi
generation_counts = data['Generation'].value_counts()

# menampilkan jumlah PokÃ©mon per generasi
print("Jumlah Pokemon per Generasi:")
print(generation_counts)

# membuat pie chart
plt.figure(figsize=(8, 8))
colors = plt.get_cmap('Set2').colors
plt.pie(generation_counts, autopct='%1.1f%%', startangle=140, colors=colors)

# menambahkan keterangan 
plt.legend(generation_counts.index, title='Generasi', loc='best', bbox_to_anchor=(1, 1))

plt.title('Persentase Pokemon Berdasarkan Generasi')
plt.axis('equal')
plt.show()
```

```python
# menghitung jumlah PokÃ©mon Legendary dan Non-Legendary
legendary_counts = data['Legendary'].value_counts()

# menampilkan jumlah PokÃ©mon berdasarkan kategori Legendary
print("Jumlah Pokemon Legendary:")
print(f"Legendary: {legendary_counts.get(True, 0)}")
print(f"Non-Legendary: {legendary_counts.get(False, 0)}")

# membuat pie chart
plt.figure(figsize=(8, 6))
colors = ['#ff9999', '#66b3ff']
plt.pie(legendary_counts, labels=['Non-Legendary', 'Legendary'], autopct='%1.1f%%', startangle=140, colors=colors)

# menambahkan keterangan
plt.legend(labels=['Non-Legendary', 'Legendary'], title='Status Legendary', loc='upper left', bbox_to_anchor=(1, 1))

plt.title('Persentase Pokomon Berdasarkan Status Legendary')
plt.axis('equal')  
plt.show()
```
